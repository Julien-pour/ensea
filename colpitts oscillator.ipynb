# colpitts oscillator

```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
```

```python
from scipy.integrate import odeint
α = 6.27; γ = 0.07; q = 0.68; η = 6.27
```

```python
def F(X,t):
    x,y,z = X[0],X[1],X[2] # ou bien x,y = X
    return α*y, -γ*(x+z)-q*y, η*(y+1-np.exp(-x))

temps = np.linspace(0,1000, 100000)
sol = odeint(F,[0.6,0.45,-0.35], temps)
```

```python
X1=sol[:,0]
X2=sol[:,1]
X3=sol[:,2]
t=temps
```

```python
plt.figure() 
plt.subplots(figsize=(14,14))
plt.subplot(3, 1, 1) # (rows, columns, panel number)
plt.title("x1=f(t)")
plt.ylabel("x1")
plt.xlabel("t en seconde")
plt.plot(t,X1)
plt.subplot(3, 1, 2) 
plt.title("x2=f(t)")
plt.ylabel("x2")
plt.xlabel("t en seconde")
plt.plot(t,X2)
plt.subplot(3, 1, 3) 
plt.title("x3=f(t)")
plt.ylabel("x3")
plt.xlabel("t en seconde")
plt.plot(t,X3)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(xs=X1, ys=X2, zs=X3,zdir='z')
plt.show()
```
